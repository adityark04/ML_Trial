# First ML Project


## Load Data

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# Data Preparation


### Data seperation of x and y

```python
y = df['logS']
y
```

```python
X = df.drop('logS' , axis=1)
X
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train , y_test = train_test_split(X, y, test_size=0.2 , random_state=100)
```

```python
X_train
```

```python
X_test
```

# Model Building

## Linear Regression

Training the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

Applying the Model to make Prediciton

```python
y_lr_train_pred = lr.predict(X_train) #predict the target values (y) for the same input data it was trained on.
y_lr_test_pred = lr.predict(X_test)  # makes predictions on the test data
y_lr_train_pred
```

```python
y_lr_test_pred
```

Evaluate Model Performance

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train , y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results = pd.DataFrame (['Linear Regression', lr_train_mse, lr_train_r2 , lr_test_mse,lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE' , ' Training R2' ,'Test MSE' , 'Test R2']
lr_results
```

## Random Forest

### Training the Model

```python
from sklearn.ensemble import RandomForestRegressor #builds multiple decision trees and combines their results
rf = RandomForestRegressor(max_depth=2, random_state=100)
#Each decision tree in the forest can grow only up to 2 levels deep
#random_state=100 says that The same "random" forest will be created every time you run the code.
rf.fit(X_train , y_train) #xtrain are features/independent variables, ytrain are target values/dependent variables
```

### Applying model to make predicition

```python
y_rf_train_pred = rf.predict(X_train) #predict the target values (y) for the same input data it was trained on.
y_rf_test_pred = rf.predict(X_test)
```

Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train , y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame (['Linear Regression', rf_train_mse, rf_train_r2 , rf_test_mse,rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE' , ' Training R2' ,'Test MSE' , 'Test R2']
rf_results
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```
